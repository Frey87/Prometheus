**Task 6.**

Create a list that will consist of 8 values ​​of valid type. Organize the input of this list from the keyboard. Swap the maximum and minimum elements of the list.

image.png

```python
mas = []
for i in range(8):
  el = float(input("Введіть значення дійсного типу  для заповнення списку:"))
  mas.append(el)

max = mas[0]
min = mas[0]
imax = 0
imin = 0

for i in range(8):
  if mas[i] > max:
    max = mas[i]
    imax = i
  if mas[i] < min:
    min = mas[i]
    imin = i

mas[imin], mas[imax] = mas[imax], mas[imin]

for i in range(8):
  print(mas[i])
```

**Task 8.**

Enter a number from the keyboard. Find the sum of the digits of a number (see problem 9 algorithms with cycles)

image.png

```python
n = int(input("Введіть число: "))
mas  = []

while n!=0:
  mas.append(n%10)
  n = n//10

sum = 0

for i in range(len(mas)):
  sum = sum + mas[i]

print(sum)
```

**Task 9.**

Write a program that forms a list of 10 integers - members of an arithmetic progression (the user specifies the value of the first term and the difference of the arithmetic progression). Find the sum of the ten terms of this arithmetic progression.

image.png

```python
mas = []
first_elem = int(input("Введіть значення першого члена: "))
step = int(input("Введіть різницю арифметичної прогресії: "))

mas.append(first_elem)

for i in range(1, 10):
  mas.append(mas[i-1]+step)

sum = 0

for i in range(10):
  sum = sum + mas[i]
print()
print(sum)
```

**Task 10.**

Check if the entered number is a palindrome

image.png

```python
n = int(input("Введіть число: "))
mas  = []

while n!=0:
  mas.append(n%10)
  n = n//10

count = len(mas)//2
flag= True

for i in range(count):
  if mas[i] != mas[len(mas)-1-i]:
    flag = False
    break
if flag == True:
  print("Palindrome")
else:
  print("Not a palindrome")
```
