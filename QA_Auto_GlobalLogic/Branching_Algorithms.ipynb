Task 6. Write a program in which the user enters a temperature value in degrees Celsius, and if this value is less than or equal to 0, the message A cold, isnt it? should be output. If the temperature is above 0 and below 10 degrees Celsius, the message will be "Cool", in other cases "Nice weather we're having".

image.png

```python
# Input: Get temperature value from the user
temp = float(input("Enter the temperature in degrees Celsius: "))

# Logic: Determine the weather message based on the temperature
if temp <= 0:
    res = "A cold, isn't it?"
elif temp < 10:
    res = "Cool"
else:
    res = "Nice weather we're having"

# Output: Display the result
print(res)
```

Task 7. Two speeds are known: one given in kilometers per hour, the other in meters per second. Which of the speeds is greater?

image.png

```python
# Input: Get two speeds from the user
speed1 = float(input("Enter the speed in kilometers per hour (km/h): "))
speed2 = float(input("Enter the speed in meters per second (m/s): "))

# Conversion: Convert speed1 from km/h to m/s
speed1_converted = speed1 * 1000 / 3600  # km/h to m/s

# Logic: Compare the speeds and determine the result
if speed1_converted > speed2:
    res = "First speed is higher"
elif speed1_converted < speed2:
    res = "Second speed is higher"
else:
    res = "Speeds are the same"

# Output: Display the result
print(res)
```

Task 8. Write a program that queries two integers. If the product of numbers exceeds their sum, print the product of numbers, otherwise print their sum. If the product is equal to the sum, output the difference of numbers.

image.png

```python
# Input: Get two integers from the user
num1 = int(input("Enter the first integer: "))
num2 = int(input("Enter the second integer: "))

# Calculate the sum and product of the numbers
sum_numbers = num1 + num2
product_numbers = num1 * num2

# Logic: Compare sum and product to determine the result
if product_numbers > sum_numbers:
    res = product_numbers
elif product_numbers < sum_numbers:
    res = sum_numbers
else:
    res = num1 - num2  # Difference of the two numbers

# Output: Display the result
print(res)
```

Task 9. The university uses the following scale to interpret student test results: 90 points and above - grade A, 80-89 points - grade B, 70-79 points - grade C, 60-69 points - grade D, below 60 - grade F. Write a program that, given the test score entered, returns the grade for that score.

image.png

```python
# Input: Get the test score from the user
mark = float(input("Enter the test score: "))

# Logic: Determine the grade based on the test score
if mark >= 90:
    res = "A"
elif mark >= 80:
    res = "B"
elif mark >= 70:
    res = "C"
elif mark >= 60:
    res = "D"
else:
    res = "F"

# Output: Display the grade
print("The grade is:", res)
```

Task 10*. Three integers are given. Determine how many of them match. The program should output one of the numbers: 2  if all the numbers are the same, 1  if the two numbers are the same, or 0  if all the entered numbers are different.

image.png

```python
# Input: Get three integers from the user
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
c = int(input("Enter the third number: "))

# Logic: Determine how many of the numbers match
if a == b == c:
    count = 2  # All three numbers are the same
elif a == b or a == c or b == c:
    count = 1  # Two numbers are the same
else:
    count = 0  # All numbers are different

# Output: Display the result
print(count)
```
