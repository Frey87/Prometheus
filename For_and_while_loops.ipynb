**Task 4.**

Given a positive integer n, calculate the value of the factorial of this number n! -.

Example: 5! = 1 · 2 · 3 · 4 · 5 = 120

image.png

```python
n = int(input())
factorial = 1
for i in range(1, n+1, 1):
    factorial = factorial*i
print(factorial)
```

**Task 6.**

Homogeneous cargo is brought to the warehouse on machines of different load capacities. The user enters information about the weight of the load of the next car. Compile a program for counting the number of cars that arrived at the warehouse before it is full, if there is capacity composition of no more than 100 tons.

image.png

```python
real_weight = 0
count = 0
weight = int(input("Enter the weight of the fisrt car:"))
real_weight = weight
while real_weight<=100:
    count += 1
    weight = int(input("Enter the weight of the next car:"))
    real_weight = real_weight + weight
print(count)
```

**Task 8.**

Write a program to print the integers from n to 0, outputting next to each number the number of # characters equal to the value of the number.

image.png

```python
n = int(input("Enter number: "))
s = "#"
count = n
while count >= 0:
    print(count, end=" ")
    count1 = 0
    while count1 < count:
        print(s, end="")
        count1 += 1
    print()
    count -= 1
```

**Task 9.**

Enter a number from the keyboard. Display all its numbers

```python
n = int(input("Enter number: "))
temp = n
div = 1
while temp >9:
    div = div *10
    temp = temp // 10
while div>0:
    print(n//div)
    n = n % div
    div = div // 10
```

**Task 10.**

Check if the entered number is prime.

image.png

```python
import math
n = int(input("Enter number: "))
flag = False
div = 2
while div<= math.sqrt(n) and not(flag):
    if n % div == 0:
        flag = True
    div += 1
if flag == True:
    print("Not a prime number")
else:
    print("Prime number")
```
